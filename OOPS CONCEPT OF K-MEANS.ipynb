# LET US IMPLEMENT OUR OWN K-MEANS ALGORITHMS

```python
# IMPLEMENTING THE BACKEND OF K-MEANS OF  HOW IT WORKS REPRESENTING IN TERMS OF CLASS 
```

```python
# LET US WORK ON TWO FACORTS ONE IS THE CLASS WITH WE CAN VISUALIZE AND ONE THE CLASS WITH ALL THE DATA WHICH IS PRESENT IN THAT
#LET US ADD TWO TYPES OF VISULAIZATION HERE AND LET US ADD PREDICT LET US ADD FIT FUNCTION INTO THIS
class Kmeans:
    
    #INITLAIZING WITH THE INIT FUNCTION 
    def __init__(self,n_clusters,max_iter,random_state):
        self.n=n_clusters
        self.max_iter=max_iter
        self.random_state=random_state
        
        
    # NOW LET US WRITE THE FIT FUNCTION
    #IN THE FIT FUNCTION WE WILL GET THE DATA
    def fit(self,data):
        
        # LET US CREATE THE RANDOM INITLIZATION OF THE CLUSTERS
        self.data=data
        knmeans=[]
        for i in range(self.n):
            newk=self.data[i]
            knmeans.append(newk)
        # YOU HAVE APPENDED THE RADOMLY INITLIAZED CLUSTERS WITH THAT
        #YOU HAVE TO ITERATE FOR MAXIMUM NUMBER OF ITERATIONS
        
        for i in range(self.max_iter):
            
            clusters=[]
            for i in range(self.n):
                clusters.append([])
                
            for point in self.data:
                #GETTING THE DISTANCE OF ALL POINTS
                alldistances=[ ((point-m)**2).sum() for m in knmeans]
                mindistance=min(alldistances)
                # GET THE INDEX OF THE MINIMUM DISTANCE
                minindex=alldistances.index(mindistance)
                clusters[minindex].append(point)
                
                
            #YOU HAVE TO AGAIN CALCULATE THE NEW CENTROID OR THE POINT OF CLUSTERS
            # YOU HAVE TO CHANGE THE knmeans values
            for i in range(self.n):
                average=np.average(clusters[i],axis=0)
                knmeans[i]=average
                
        #JUST RETURN THE knmeans 
        return np.array(knmeans)
    
    # NOW LET US WRITE THE PREDICTION FUNCTION
    # YOU HAVE TO PREDICT FUNCTION WHICH TAKES data and Values of the clusters
    def predict(self,data,meanvalues):
        self.data=data
        self.values=meanvalues
        
        predictions=[]
        
        for point in self.data:
            alldistances=[((point-m)**2).sum() for m in self.values]
            mindistance=min(alldistances)
            minindex=alldistances.index(mindistance)
            predictions.append(minindex)
        return np.array(predictions)
    
    #LET US VISULIZE THE DATA HERE
    def visualize(self,data,powermeans,predictions):
        self.data=data
        self.powermeans=powermeans
        self.predictions=predictions
        
        #DRAW THE PLOT
        
        plt.scatter(self.data[:,0],self.data[:,1],cmap=self.predictions)
        plt.scatter(self.powermeans[:,0],self.powermeans[:,1])
        plt.axis("OFF")
        plt.xlabel("Annual Income (k$)")
        plt.ylabel("Spending Score (1-100)")
        plt.title("VISULISATION")
        plt.show()
        
    def neatvisualize(self,data,powermeans,predictions):
        self.data=data
        self.powermeans=powermeans
        self.predictions=predictions
        
        # DRAW THE PLOT
        plt.scatter(self.data[self.predictions==0,0],self.data[self.predictions==0,1],s=100,c='red',label='Cluster 1')
        plt.scatter(self.data[self.predictions==1,0],self.data[self.predictions==1,1],s=100,c='blue',label='Cluster 2')
        plt.scatter(self.data[self.predictions==2,0],self.data[self.predictions==2,1],s=100,c='green',label='Cluster 3')
        plt.scatter(self.data[self.predictions==3,0],self.data[self.predictions==3,1],s=100,c='yellow',label='Cluster 4')
        plt.scatter(self.data[self.predictions==4,0],self.data[self.predictions==4,1],s=100,c='cyan',label='Cluster5')
        plt.scatter(self.powermeans[:,0],self.powermeans[:,1],s=300,c='magenta')
        plt.title('PERFECT VISULISATION')
        plt.xlabel('Annual Income (k$)')
        plt.ylabel('Spending Score (1-100)')
        plt.show() 
```

```python
ckmeans=Kmeans(5,100,0)
```

```python
powermeans=ckmeans.fit(X)
```

```python
print(X)
```

```python
poweryes=ckmeans.predict(X,powermeans)
```
